 A Obtain the 50th Fibonacci number

```python
; multi-segment executable file template.
org 100h

jmp start


NUM1 dw 1
NUM2 dw 1

START:
  
mov 	cx, 50 
mov 	ax, NUM1
ends   

Fibonacci:
    mov 	bx, NUM1
    add 	bx, NUM2
    mov 	dx, bx
    mov 	bx, NUM1
    mov 	NUM2, bx
    mov 	bx, dx
    mov 	NUM1, bx

    

loop Fibonacci  

mov ah, 0
int 16h
ret 

end start
```
